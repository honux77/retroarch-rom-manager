## 환경 변수 불러오기


```python
romdir = "D:\\Retropie\\roms"
import os
os.chdir(romdir)
subdirs = [d for d in os.listdir(romdir) if os.path.isdir(os.path.join(romdir, d))]
print(subdirs)
```

```python
# find manual directories in subdirs
# remove them if they exist
manuals = []
import shutil
for subdir in subdirs:
    manual_path = os.path.join(subdir, "media\\manual")
    if os.path.exists(manual_path):
        shutil.rmtree(manual_path)
        print(f"Removed manual directory: {manual_path}")

```
